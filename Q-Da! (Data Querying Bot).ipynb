# Q-Da : Your friendly neighborhood data query bot!

```python
import pandas as pd 


error = "Sorry, this is invalid. Please type 'help' for more info."

verbs = ["get","set","loadcsv","list","exit", "easter", "apply", "help"]
shapes = ["row","column","cell", "df", "egg", "columns", "rlength", "numerical", "filter"]

def parse_text(text):
    '''returns text separated by space'''
    text_list = text.split()
    text_phrase = {"verb":None,"shapes":None,"var":None}
    
    for i in verbs: 
        if i in text_list:
            if text_phrase["verb"] != None:
                print("A statement should only have one verb")
                return None
            text_phrase["verb"] = i
            text_list.pop(text_list.index(i))
    
    for i in shapes: 
        if i in text_list:
            if text_phrase["shapes"] != None:
                print("A statement should only have one verb")
                return None
            text_phrase["shapes"] = i
            text_list.pop(text_list.index(i))
    
    if len(text_list) == 1: 
        text_phrase["var"] = text_list[0]
    if len(text_list) == 2:
        text_phrase["var1"] = text_list[0]
        text_phrase["var2"] = text_list[1]
  
    
    return text_phrase
    


#get row with row number as argument
def get_row(df, row):
    row = int(row)
    if row <= len(df):
        return pd.DataFrame(df.iloc[row])
    else:
        return error

#get column with column name as argument
def get_col(df, col):
    if col in list(df.columns):
        ans = pd.DataFrame(df[col])
        return ans 
    else:
        return error
#list all columns in dataframe
def list_cols(df):
    
    return df.columns
#get a spcific cell of dataframe 
def get_cell(df, cell):
    
    if len(cell) == 2:
        cell_row = int(cell[0])
        cell_column = int(cell[1])
        return df.iloc[cell_row, cell_column]
    elif len(cell) >= 2:
        cell_row = int(cell[:(len(cell)-2)])
        cell_column = int(cell[-1])
        return df.iloc[cell_row, cell_column]
    else:
        return error
#change/set a cell value  
def set_cell(df, cell, val):
    
    if len(cell) == 2:
        cell_row = int(cell[0])
        cell_column = int(cell[1])
        df.iloc[cell_row, cell_column] = str(val)
    elif len(cell) >= 2:
        cell_row = int(cell[:(len(cell)-2)])
        cell_column = int(cell[-1])
        df.iloc[cell_row, cell_column] = str(val)
    else:
        return error
    print(f"The value of the cell has been set to {val}")
    
#get length of a row in dataframe  
def get_rlength(df):
    return len(df.columns)

##extra function 1:
#get all the numerical columns in dataset
def get_numerical(df):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

    newdf = df.select_dtypes(include=numerics)
    sq = input("'Would you like to save this dataframe? (y/n)'")
    if sq == "y":
        name = "Numerical_df.xlsx"
        newdf.to_excel(name)
        print(f"Numerical Columns have been saved as {name}")
    else:
        print("Numerical columns have not been saved. They are as follows:")
    return newdf

##extra function 2:
#filter columns using numerical value
def filter_col(df, col, val):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    val = int(val)
    if col in df.select_dtypes(include=numerics).columns :
        compare = ["<",">","!=", "==", "<=", ">="]
        comp = input('"Please enter one of "<",">","!=", "==", "<=", ">="')
        if comp == "<":
            df_filtered = df[df[col] < val]
        if comp == ">":
            df_filtered = df[df[col] > val]
        if comp == "!=":
            df_filtered = df[df[col] != val]
        if comp == "==":
            df_filtered = df[df[col] == val]
        if comp == "<=":
            df_filtered = df[df[col] <= val ]
        if comp == ">=":
            df_filtered = df[df[col] >= val ]
        sq = input("'Would you like to save this dataframe? (y/n)'")
        if sq == "y":
            name = "Filtered_df.xlsx"
            df_filtered.to_excel(name)
            print(f"NFiltered Rows have been saved as {name}")
        else:
            print("Filtered columns have not been saved. They are as follows.")
        return df_filtered
    else:
        print("The column entered is not numerical. Try 'get numerical' for more info.")
        
#Easter egg function to print jokes
def easter_egg():
    print('What is red coloured, that smells like blue paint?')
    input()
    print('Red paint!')
    print()
    print('What do you call a flying bagel?')
    input()
    print('A plain (PLANE) bagel!')
    
    
    
    

running = True
while True: 
    print('Welcome to Q-Da!\n\
          help : find out more about actions available in Q-Da! \n\
          exit : to leave')
         

    text = input()
    if text == "": 
        continue
    
    actions = parse_text(text)
    if not actions: 
        print("Please try again or write type 'exit' to leave")
        continue
    
    #print(actions)
    
    if actions["verb"] == "exit":
        break
    elif actions["verb"] == "loadcsv":
        df = pd.read_csv(actions["var"])
        print(f"The DataFrame is loaded. It has {len(df)} rows")
    elif actions["verb"] == "get" and actions["shapes"] == "row":
            print(get_row(df, actions["var"]))
    elif actions["verb"] == "get" and actions["shapes"] == "column":
            print(get_col(df, actions["var"]))
    elif actions["verb"] == "list" and actions["shapes"] == "columns":
            print(list_cols(df))
    elif actions["verb"] == "get" and actions["shapes"] == "cell":
            print(get_cell(df, actions["var"]))
    elif actions["verb"] == "set" and actions["shapes"] == "cell":
            print(set_cell(df, actions["var1"], actions["var2"]))
    elif actions["verb"] == "get" and actions["shapes"] == "rlength":
            print(get_rlength(df))
    elif actions["verb"] == "get" and actions["shapes"] == "numerical":
            print(get_numerical(df))
    elif actions["verb"] == "apply" and actions["shapes"] == "filter":
            print(filter_col(df, actions["var1"], actions["var2"]))
    elif actions["verb"] == "easter" and actions["shapes"] == "egg":
            print(easter_egg())
    elif actions["verb"] == "help":
        print("'We offer the following commands : \n\n\n\
         loadexcel <name of csv>: loads the dataset, indicates number of rows\n\
         get column <column name>: to extract a particular column\n\
         get row <row number>: to extract a particular row\n\
         list columns: to list a column\n\
         get cell <RowNumberColumnNumer> (without spaces): to extract a particular cell\n\
         set cell <RowNumberColumnNumber> (without spaces) <Value to set in cell>: to change/set a cell value \n\
         get rlength: to get length of a row \n\
         get numerical : to get all the columns with numerical values in dataset, with option to save this as excel \n\
         apply filter <column to filter> <value to filter by> : apply numerical filter to any numerical column in dataset,with an option to choose which inequality/equility filter to apply \n\
         easter egg : try it to find out! \n\
         exit : to leave \n\'")
    

# Note: loadcsv ../data/imdb_over10k_votes_movies.csv 
#       will work if your data is in a sibling folder called 'data'
```
